```python
df= pd.read_csv('/kaggle/input/obesity-levels/ObesityDataSet_raw_and_data_sinthetic.csv')
df.head()
```

```python
cols = df.columns
num_cols = df._get_numeric_data().columns
char_cols=list(set(cols) - set(num_cols))
char_cols
```

```python
df.describe()
```

```python
df.isna().count()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
from sklearn.impute import SimpleImputer
si=SimpleImputer()
def encode(df):
    for i in char_cols:
        df[i]= le.fit_transform(df[i])
    return df
def impute(df):
    df[num_cols]=si.fit_transform(df[num_cols])
    return df

    
```

```python
encode(df)
```

```python
impute(df)
```

```python
from sklearn.model_selection import train_test_split
X= df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,stratify=df[['NObeyesdad']])
```

```python
from xgboost import XGBClassifier
clf=XGBClassifier()
clf.fit(X_train,y_train)
y_pred= clf.predict(X_test)
from sklearn.metrics import f1_score
f1_score(y_test, y_pred, average='weighted')
```
